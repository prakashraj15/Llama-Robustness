```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DATA.csv')
x = data.iloc[:, 0].values
y = data.iloc[:, 1].values
```

```python
print(x) 
print(y)

```

```python
plt.figure(figsize=(8, 5))
plt.bar(x, y, color='skyblue')
plt.xlabel('Targets')
plt.ylabel('Values')
plt.title('Liberal Toxicity')
plt.yscale('log')
plt.ylim(1e-4,1)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
